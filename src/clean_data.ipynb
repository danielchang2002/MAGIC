# Clean pathway coverage

```python
pathcoverage = pd.read_csv(os.path.join(config.RAW_DIR, "humann_pathcoverage_overall_reduced.tsv"), sep="\t", index_col=0).T
pathcoverage
```

```python
pattern = re.compile("(^magic)|(^[0-9])")

non_controls = [ind for ind in pathcoverage.index if pattern.match(ind)]
non_controls[:10]
```

```python
non_control_df = pathcoverage.loc[non_controls]
non_control_df["id"] = [ind.replace("-", ".").replace("_", ".")[:10] if ind.startswith("magic") else "magic." + ind.split("_")[0].zfill(4) for ind in non_control_df.index]
non_control_df
```

```python
drop_filenames = []

for duplicate in duplicates:
    filename = duplicate[:-6].replace(".", "_") + "Coverage"
    if filename[0] == "X":
        filename = filename[1:]
        
    if filename in non_control_df.index:
        drop_filenames.append(filename)
    else:
        filename = "magic-" + filename[6:]
        if filename in non_control_df.index:
            drop_filenames.append(filename)
        else:
            print("bruh")
            break
len(drop_filenames)    
```

```python
path_coverage_clean_with_filenames = non_control_df.drop(drop_filenames)
path_coverage_clean = path_coverage_clean_with_filenames.set_index("id").sort_index()
path_coverage_clean
```

```python
path_coverage_clean.to_csv(os.path.join(config.CLEAN_DIR, "humann_pathcoverage_clean.tsv"), sep="\t")
```
