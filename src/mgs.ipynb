# MGS

```python
mgs = pd.read_csv(os.path.join(config.RAW_DIR, "humann_reduced_mgs.tsv"), sep="\t", index_col=0, header=None).iloc[:, :-1]
header = pd.read_csv(os.path.join(config.CLEAN_DIR, "humann_pathcoverage_overall_clean.tsv")).columns[1:]
mgs.columns = header
mgs
```

```python
plt.figure(figsize=(10, 5), dpi=200)

mgs_presence = (mgs > 0)
counts = mgs_presence.sum(axis=0)

plt.hist(counts, bins=75, edgecolor="black")
plt.xlabel("\# of CAGs present in a given sample", fontsize=20)
plt.ylabel("\# of samples", fontsize=20)

plt.show()
```
